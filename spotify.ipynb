```python
#scatter plot
import matplotlib.pyplot as plt
X = data['id']
Y = data['duration_ms']
plt.scatter(x = X , y = Y,marker = '*',c='green',alpha=0.2)
plt.scatter(x = X , y = Y,marker = '*',c='red',alpha=0.2,linewidths=10)
plt.title("id v/s duration_ms")
plt.xlabel("id")
plt.ylabel("duration_ms")
plt.grid()
plt.show()
```
